```python
import math
import networkx as nx
import random
import matplotlib.pylab as plt
%matplotlib inline
```

```python
G = nx.Graph()
# G.add_node(1) das ist eine Node mit der ID 1. Jede Node braucht eine ID
# G.add_node("Thomas") Hier ist die ID ein String. Merke dir: die ID kann alles sein, Zahl oder String!

num_nodes= 100
last_node= None

for i in range(num_nodes):
    x = random.randint(0,100)
    y= random.randint(0,100)

    G.add_node(i, pos=(x,y)) # Hier wird die ID i als Node-ID verwendet. Die Position ist ein Tuple mit x und y Koordinaten
    # TASK 2: a better solution, a complete graph is too messy
    """
    if last_node is not None:
        G.add_edge(last_node, i)
    last_node= i
    """
    

#print("Nodes of G are: {}".format(G.nodes()))
```

```python
G.nodes(data=True)
```

```python
# Connect nodes within a certain distance
distance= 10

for i in range(num_nodes):
    for j in range(i+1, num_nodes):
        pos_i= G.nodes[i]['pos']
        pos_j= G.nodes[j]['pos']
        dist= math.sqrt((pos_i[0]-pos_j[0])**2 + (pos_i[1]-pos_j[1])**2)
        if dist < distance:
            G.add_edge(i, j)
            print("Adding edge between {} and {}".format(i,j))
```

```python
# make complete graph
complete_G= nx.complete_graph(num_nodes) # define the number of nodes
# give the coordinates of the nodes from the above genrrated graph
for node in complete_G.nodes():
    complete_G.nodes[node]['pos'] = G.nodes[node]['pos']
```

```python
pos = nx.get_node_attributes(G, 'pos')
pos
```

```python
nx.draw(G, pos=pos, with_labels=True, node_color="lightgreen")
plt.show()
```
